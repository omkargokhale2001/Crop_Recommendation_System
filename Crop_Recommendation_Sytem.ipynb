### Model Training

#### KNN Classifier

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
le = LabelEncoder()
scaler = MinMaxScaler()
y = df['label']
X = df.drop(['label'],axis=1)
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier()
grid_params = {'n_neighbors':[1,3,5,7,9]}
grid = GridSearchCV(estimator = model_knn,param_grid=grid_params)
grid.fit(X,y)
model1 = KNeighborsClassifier(n_neighbors=grid.best_estimator_.n_neighbors).fit(X_train_scaled,y_train)
score_scaled = model1.score(X_test_scaled,y_test)
model2 = KNeighborsClassifier(n_neighbors=grid.best_estimator_.n_neighbors).fit(X_train,y_train)
score = model2.score(X_test,y_test)
score,score_scaled
```

```python
# Predicting a sample value:
example = X.iloc[500]
lst=[]
df['predict_label'] = model1.predict(scaler.transform(X))
ans = df['predict_label'].value_counts()
for i in example:
    lst.append(i)
data1 = np.array(lst).reshape(1,-1)
data1 = scaler.transform(data1)
(model1.predict(data1),y.iloc[500])
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_logic = LogisticRegression(C=1000,max_iter=1000,multi_class='multinomial',solver='lbfgs').fit(X_train_scaled,y_train)
# grid_params={'C':[0.01,0.1,1,10,100,1000]}
# grid = GridSearchCV(estimator=model_logic,param_grid=grid_params)
# grid.fit(X,y)
# grid.best_estimator_.C
score = model_logic.score(X_test_scaled,y_test)
score
```

#### Support Vector Machine

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

```python
model_svc = SVC()
grid_params = {'kernel':['linear','rbf','poly'],'degree':[1,2,3,4,5]}
grid_svc = GridSearchCV(estimator=model_svc,param_grid=grid_params)
grid_svc.fit(X_train_scaled,y_train)
kernel_best,degree_best = grid_svc.best_estimator_.kernel,grid_svc.best_estimator_.degree
model_svc=SVC(kernel=kernel_best,degree = degree_best).fit(X_train_scaled,y_train)
score = model_svc.score(X_test_scaled,y_test)
score
```

```python
# Predicting a sample value
example = X.iloc[500]
lst=[]
df['predict_label'] = model_svc.predict(scaler.transform(X))
ans = df['predict_label'].value_counts()
for i in example:
    lst.append(i)
data1 = np.array(lst).reshape(1,-1)
data1 = scaler.transform(data1)
(model_svc.predict(data1),y.iloc[500])
```

#### Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.inspection import permutation_importance
```

```python
model_tree = DecisionTreeClassifier()
grid_params = {'max_depth':[1,2,3,4,5,6,7,8,9],'max_leaf_nodes':[2,3,4,5,6,7,8,9]}
grid_tree = GridSearchCV(estimator=model_tree,param_grid=grid_params)
grid_tree.fit(X_train_scaled,y_train)
model_tree = DecisionTreeClassifier().fit(X_train_scaled,y_train)
model_tree.score(X_test_scaled,y_test)
```

```python
#Important features
importances = permutation_importance(model_tree,X_test_scaled,y_test)
df_feature_importance = pd.DataFrame(importances.importances_mean,index=X.columns)
df_feature_importance
```

```python
# Example
example = X.iloc[500]
lst=[]
df['predict_label'] = model_tree.predict(scaler.transform(X))
ans = df['predict_label'].value_counts()
for i in example:
    lst.append(i)
data1 = np.array(lst).reshape(1,-1)
data1 = scaler.transform(data1)
(model_tree.predict(data1),y.iloc[500])
```
